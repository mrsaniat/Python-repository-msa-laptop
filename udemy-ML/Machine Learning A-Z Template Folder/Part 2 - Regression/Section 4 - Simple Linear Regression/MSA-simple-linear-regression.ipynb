```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#importing dataset
dataset = pd.read_csv('Salary_Data.csv')

#show dataset
print(dataset)

#create independent variable vector (country age salary)
X = dataset.iloc[:,:-1].values
X1 = dataset.iloc[:,:-1]


#create dependent variable vector (Purchased)
Y = dataset.iloc[:,1].values
Y1 = dataset.iloc[:,1]

#check what is in the X and X1:
print('\n ---- this is the seperator ---- \n')
print("value of X:",X)

print('\n ---- this is the seperator ---- \n')
print("value of X1:",X1)

#create dependent variable now
#Y = dataset.iloc[:,3].values
print('\n ---- this is the seperator ---- \n')
print("value of Y:",Y)


from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.33, random_state = 0)

print(" \n X_train: \n", X_train)
print(" \n Y_train: \n", Y_train)
print(" \n X_test: \n", X_test)
print(" \n Y_test: \n", Y_test)
```

```python
### Apply simple linear regression ###
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)

### predict
Y_pred = regressor.predict(X_test)
#Saniat: convert the float into number
#import math
#Y_predicted = math.floor(Y_pred)
print("\n real value Y_test:",Y_test)
print("\n predicted value Y_pred:",Y_pred)
```

```python
### visualising simple linear regression
plt.scatter(X_train,Y_train, color = 'red')
#
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
